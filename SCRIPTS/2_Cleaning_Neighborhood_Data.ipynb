##This script cleans each individual neighborhood dataset. There are 3 datasets total, and each has data on 2 neighborhoods. So, we have data from 6 neighborhoods in total. Repeat this process for each neighborhood.

```python
# Import necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import re
```

```python
# Read the CSV file
# Repeat for each uncleaned data csv
df=pd.read_csv('DS4002/DATA/Uncleaned Data/Uncleaned_JeffersonParkFry Spring.csv')
```

```python
df.head()
```

```python
# Check data types
df.dtypes
```

```python
# Convert date from object data type to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d', errors='coerce')
```

```python
df.dtypes
```

```python
# Convert % into decimal by dividing by 100
columns_to_convert = ['Median Sale Price MoM', 'Median Sale Price YoY', 'Homes Sold MoM', 'Homes Sold YoY']
df[columns_to_convert] = df[columns_to_convert].div(100).round(2)
```

```python
df.head()
```

```python
# Convert cleaned data to CSV
df.to_csv('Cleaned_JeffParkFrySpring.csv', index=False)
```
