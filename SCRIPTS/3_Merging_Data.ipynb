##This notebook merges the Charlottesville regional data with the 6 selected neighborhood data.

```python
# Import necessary packages
import pandas as pd
import numpy as np
```

```python
# Import cleaned CSVs
df1=pd.read_csv('DS4002/DATA/Cleaned Data/ Cleaned_Charlottesville.csv')
df2=pd.read_csv('DS4002/DATA/Cleaned Data/ Cleaned_FifevilleMarthaJefferson.csv')
df3=pd.read_csv('DS4002/DATA/Cleaned Data/Cleaned_JeffersonParkFrySpring.csv')
df4=pd.read_csv('DS4002/DATA/Cleaned Data/Cleaned_VenableBelmont.csv')
```

```python
# Combine cleanded data
merged_df = pd.concat([df1, df2, df3, df4], ignore_index=True)
```

```python
merged_df.head(30)
```

```python
# organize by ascending date
df_sorted = merged_df.sort_values(by='Date', ascending=True)
```

```python
df_sorted.head(20)
```

```python
# Convert to csv!
df_sorted.to_csv('Cleaned_Final_Data.csv', index=False)
```
