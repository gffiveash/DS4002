##This notebook cleans the larger Charlottesville dataset, removing unnecessary columns, formatting with Datetime, and cutting the dataset.

```python
# Connect to GitHub
! git clone https://github.com/gffiveash/DS4002
```

```python
# Import necessary packages
import pandas as pd
```

```python
# Read the CSV file
df=pd.read_csv('DS4002/DATA/Uncleaned Data/Uncleaned_RedfinCharlottesvilleHousingMarket.csv')
```

```python
df.head()
```

```python
# Removes the unnecessary columns
df = df.drop(['New Listings','New Listings MoM','New Listings YoY','Inventory','Inventory MoM','Inventory YoY','Days on Market','Days on Market MoM','Days on Market YoY','Average Sale To List','Average Sale To List MoM','Average Sale To List YoY'], axis=1)
```

```python
# Df.head() to visualize dropped columns
df.head()
```

```python
# Renaming columns
df.rename(columns={'Region': 'Location', 'Month of Period End': 'Date'}, inplace=True)
```

```python
# Df.head() to visualize renamed columns
df.head()
```

```python
# Swapping order of columns so Date is first

columnorder = list(df.columns)
columnorder[0], columnorder[1] = columnorder[1], columnorder[0]
df = df[columnorder]
```

```python
# Df.head() to visualize reordered columns
df.head()
```

```python
# Using datetime to format Date column

df['Date'] = pd.to_datetime(df['Date'], format='%B %Y', errors='coerce')
```

```python
# Removing all rows before January 2020 (because neighborhood datasets start in January 2020)
df = df[df['Date'] >= '2020-01-01']
```

```python
df.head()
```

```python
# Removing string values ($, K, %, etc.) and ensuring Sales Price is an integer

df['Median Sale Price'] = df['Median Sale Price'].str.replace('$', '').str.replace('K', '').astype(int)
```

```python
# Removing string values ($, K, %, etc.) and ensuring columns are floats

columnswithpercentages = ['Median Sale Price MoM', 'Median Sale Price YoY', 'Homes Sold MoM', 'Homes Sold YoY']
for each in columnswithpercentages:
    df[each] = df[each].str.replace('%', '').astype(float) / 100
```

```python
df['Median Sale Price MoM'] = df['Median Sale Price MoM'].round(2)
df['Median Sale Price YoY'] = df['Median Sale Price YoY'].round(2)
df['Homes Sold MoM'] = df['Homes Sold MoM'].round(2)
df['Homes Sold YoY'] = df['Homes Sold YoY'].round(2)
```

```python
df.head()
```

```python
# Checking data types
df.dtypes
```

```python
# Convert to csv
df.to_csv('Cleaned_CharlottesvilleData.csv', index=False)
```
